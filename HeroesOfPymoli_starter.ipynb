```python
# Dependencies and Setup
import pandas as pd
import numpy as np
from IPython.display import display, HTML

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

```python
#Display the total number of players
player_count = len(purchase_data["SN"].unique())
player_count
total_players = pd.DataFrame({"Total Players": [player_count]})
total_players
```

Purchasing Analysis (Total)
 
    Run basic calculations to obtain number of unique items, average price, etc.
    Create a summary data frame to hold the results
    Optional: give the displayed data cleaner formatting
    Display the summary data frame

```python
unique_items_count = len(purchase_data["Item ID"].unique())
average_purchase = purchase_data["Price"].mean()
purchase_count = len(purchase_data)
revenue = purchase_data["Price"].sum()
```

```python
purchasing_analysis = pd.DataFrame({'Number of Unique Items': [unique_items_count],
                                   'Average Price': [average_purchase],
                                   'Number of Purchases': [purchase_count],
                                   'Total Revenue': [revenue]})
purchasing_analysis['Average Price'] = purchasing_analysis['Average Price'].map("${0:,.2f}".format) 
purchasing_analysis['Total Revenue'] = purchasing_analysis['Total Revenue'].map("${0:,.2f}".format) 
#display(HTML(df2.to_html()))
display(purchasing_analysis)
```

Gender Demographics
 
    Percentage and Count of Male Players
    Percentage and Count of Female Players
    Percentage and Count of Other / Non-Disclosed

```python
gender_count_group = purchase_data.groupby(['Gender'])
gender_count_df = pd.DataFrame({'Total Count': gender_count_group["SN"].nunique()})
gender_count_df['Percentage of Players'] = (gender_count_df['Total Count']/player_count)*100
gender_count_df['Percentage of Players'] = gender_count_df['Percentage of Players'].map("{0:,.2f}%".format)
gender_count_df = gender_count_df.sort_values(['Total Count'],ascending=False)
gender_count_df = gender_count_df.rename_axis(None)
gender_count_df
```

Purchasing Analysis (Gender)
 
    Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender
    Create a summary data frame to hold the results
    Optional: give the displayed data cleaner formatting
    Display the summary data frame

```python
gender_per_person_df = pd.DataFrame({'Total Count': gender_count_group["SN"].nunique()})
gender_per_person_df = gender_per_person_df.reset_index()
female_count_df = gender_per_person_df.loc[gender_per_person_df['Gender'] == 'Female']
male_count_df =   gender_per_person_df.loc[gender_per_person_df['Gender'] == 'Male']

female_number = female_count_df['Total Count'][0]
male_number = male_count_df['Total Count'][1]
others_number = player_count - female_number - male_number
```

```python
purchase_by_gender = pd.DataFrame({'Purchase Count': gender_count_group['Price'].count(),
                                   'Average Purchase Price': gender_count_group['Price'].mean(),
                                   'Total Purchase Value': gender_count_group['Price'].sum(),
                                    'Gender_count': [female_number, male_number, others_number]})
purchase_by_gender['Avg Total Purchase per Person'] = (purchase_by_gender['Total Purchase Value'] / purchase_by_gender['Gender_count'])
purchase_by_gender['Average Purchase Price'] = purchase_by_gender['Average Purchase Price'].map("${0:,.2f}".format) 
purchase_by_gender['Total Purchase Value'] = purchase_by_gender['Total Purchase Value'].map("${0:,.2f}".format)
purchase_by_gender['Avg Total Purchase per Person'] = purchase_by_gender['Avg Total Purchase per Person'].map("${0:,.2f}".format)
del purchase_by_gender['Gender_count']
purchase_by_gender
```

Age Demographics
 
    Establish bins for ages
    Categorize the existing players using the age bins. Hint: use pd.cut()
    Calculate the numbers and percentages by age group
    Create a summary data frame to hold the results
    Optional: round the percentage column to two decimal points
    Display Age Demographics Table

```python
age_count_group = purchase_data.groupby(['Age'])
age_per_person_df = pd.DataFrame({'Total Count': age_count_group["SN"].nunique()})
age_per_person_df = age_per_person_df.reset_index()
```

```python
age_bin =    [0,      9,      14,      19,     24,      29,       34,      39,   200]
age_groups = ['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+']
age_per_person_df['Age Group'] = pd.cut(age_per_person_df['Age'], age_bin, labels=age_groups)
age_group_count_df = pd.DataFrame(age_per_person_df.groupby('Age Group')['Total Count'].sum())
age_group_count_df = age_group_count_df.rename_axis(None)
age_group_count_df['Percentage of Players'] = (age_group_count_df['Total Count']/player_count)*100
age_group_count_df['Percentage of Players'] = age_group_count_df['Percentage of Players'].map("{0:,.2f}".format)
age_group_count_df

```

Purchasing Analysis (Age)
 
    Bin the purchase_data data frame by age
    Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table         below
    Create a summary data frame to hold the results
    Optional: give the displayed data cleaner formatting
    Display the summary data frame

```python

purchases_df = pd.DataFrame({'Purchase Count': purchase_data.groupby("Age").size(),
                            'Total Purchase Price': purchase_data.groupby("Age")['Price'].sum(),
                            'Total People': purchase_data.groupby("Age")['SN'].nunique()})
purchases_df = purchases_df.reset_index()
purchases_df['Age Group'] = pd.cut(purchases_df['Age'], age_bin, labels=age_groups)
purchases_analysis_df = pd.DataFrame(purchases_df.groupby('Age Group')[['Purchase Count', 'Total Purchase Price', 'Total People']].sum())
purchases_analysis_df['Average Purchase Price'] = purchases_analysis_df['Total Purchase Price'] / purchases_analysis_df['Purchase Count'] 
purchases_analysis_df['Avg Total Purchase per Person'] = purchases_analysis_df['Total Purchase Price'] / purchases_analysis_df['Total People']
del purchases_analysis_df['Total People']
summary_df = purchases_analysis_df[['Purchase Count', 'Average Purchase Price', 'Total Purchase Price',  'Avg Total Purchase per Person']]
summary_df = summary_df.sort_values(['Age Group'],ascending=False)
summary_df['Average Purchase Price'] = summary_df['Average Purchase Price'].map("${0:,.2f}".format)
summary_df['Total Purchase Price'] = summary_df['Total Purchase Price'].map("${0:,.2f}".format)
summary_df['Avg Total Purchase per Person'] = summary_df['Avg Total Purchase per Person'].map("${0:,.2f}".format)
summary_df
```

Top Spenders
 
    Run basic calculations to obtain the results in the table below
    Create a summary data frame to hold the results
    Sort the total purchase value column in descending order
    Optional: give the displayed data cleaner formatting
    Display a preview of the summary data frame

```python
spender_df = pd.DataFrame({'Purchase Count': purchase_data.groupby("SN")['Price'].count(),
                           'Total Purchase Value': purchase_data.groupby("SN")['Price'].sum()
                          })
spender_df = spender_df.sort_values(['Total Purchase Value'],ascending=False)
spender_df['Average Purchase Price'] = spender_df['Total Purchase Value'] / spender_df['Purchase Count']  
spender_df['Total Purchase Value'] = spender_df['Total Purchase Value'].map("${0:,.2f}".format)
spender_df['Average Purchase Price'] = spender_df['Average Purchase Price'].map("${0:,.2f}".format)
spender_df.head()
```

Most Popular Items
 
    Retrieve the Item ID, Item Name, and Item Price columns
    Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value
    Create a summary data frame to hold the results
    Sort the purchase count column in descending order
    Optional: give the displayed data cleaner formatting
    Display a preview of the summary data frame

```python
item_summary_df = purchase_data[['Item ID','Item Name','Price']]
item_df = pd.DataFrame({'Purchase Count': item_summary_df.groupby(['Item ID', 'Item Name'])['Price'].count(),
                        'Item Price': item_summary_df.groupby(['Item ID', 'Item Name'])['Price'].first(),
    'Total Purchase Value': item_summary_df.groupby(['Item ID', 'Item Name'])['Price'].sum()})
item_df = item_df.sort_values(['Purchase Count'],ascending=False)
item_df['Item Price'] = item_df['Item Price'].map("${0:,.2f}".format)
item_df['Total Purchase Value'] = item_df['Total Purchase Value'].map("${0:,.2f}".format)
item_df.head()
```
